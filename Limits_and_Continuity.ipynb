## Introduction to Limits

### What is a Limit?

A **limit** describes how a function behaves as its input approaches a certain value. Formally,  
$$
\lim_{x \to a} f(x) = L
$$
means that as $ x $ gets closer to $ a $, $ f(x) $ approaches $ L $.

#### Example

Let’s consider $ f(x) = \frac{\sin x}{x} $ as $ x \to 0 $.

### Calculating Limits Numerically

Let’s use NumPy to approximate the limit numerically:

````python
import numpy as np

x_values = np.linspace(0.01, 1, 100)
y_values = np.sin(x_values) / x_values

print("Approximate value as x approaches 0:", y_values[0])
````

### Visualizing Limits

Plot the function near $ x = 0 $:

````python
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 400)
y = np.sinc(x/np.pi)  # np.sinc(x) = sin(pi*x)/(pi*x)

plt.plot(x, y, label='sin(x)/x')
plt.axvline(0, color='gray', linestyle='--')
plt.title('Plot of sin(x)/x near x=0')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
````

#### Calculating Symbolic Limits

With SymPy, you can compute the exact limit:

````python
from sympy import Symbol, sin, limit

x = Symbol('x')
lim = limit(sin(x)/x, x, 0)
print("Symbolic limit as x approaches 0:", lim)
````

## Exploring Continuity

### What Makes a Function Continuous?

A function $ f(x) $ is **continuous** at $ x=a $ if:

1. $ f(a) $ is defined,
2. $ \lim_{x \to a} f(x) $ exists,
3. $ \lim_{x \to a} f(x) = f(a) $

Let’s check continuity for:

$$
f(x) = 
\begin{cases}
  x^2 & x \neq 2 \\
  5 & x = 2
\end{cases}
$$

### Checking Continuity with Python

````python
from sympy import Piecewise

x = Symbol('x')
f = Piecewise((x**2, x != 2), (5, x == 2))

# Evaluate f(2)
f_at_2 = f.subs(x, 2)

# Limit as x approaches 2
lim_at_2 = limit(f, x, 2)

print("f(2) =", f_at_2)
print("lim_{x->2} f(x) =", lim_at_2)

if f_at_2 == lim_at_2:
    print("f(x) is continuous at x=2")
else:
    print("f(x) is NOT continuous at x=2")
````

## Summary

- **Limits** describe how functions behave near specific points.
- **Continuity** requires that a function’s value and its limit agree at a point.
- Python (NumPy, Matplotlib, SymPy) makes it easy to explore these concepts numerically, visually, and symbolically.
