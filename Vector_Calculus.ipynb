## 2. Vector Fields

A **vector field** assigns a vector to each point in space.  
Example in 2D:  
$$
\vec{F}(x, y) = (P(x, y), Q(x, y))
$$

### Example: Plotting a Vector Field

````python
import numpy as np
import matplotlib.pyplot as plt

x, y = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))
u = -y
v = x

plt.quiver(x, y, u, v)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Vector Field: F(x, y) = (-y, x)')
plt.axis('equal')
plt.show()
````

## 3. Line Integrals

The **line integral** of a vector field along a curve $ C $ is:
$$
\int_C \vec{F} \cdot d\vec{r}
$$

### Example: Compute the Line Integral

Let $ \vec{F}(x, y) = (y, x) $, $ C $ is the line from (0,0) to (1,1):


````python
import sympy as sp

t = sp.symbols('t')
x = t
y = t
F = sp.Matrix([y, x])
r_prime = sp.Matrix([sp.diff(x, t), sp.diff(y, t)])  # dr/dt
integrand = F.dot(r_prime)

line_integral = sp.integrate(integrand, (t, 0, 1))
print(line_integral)  # Output: 2/3
````

## 4. Green’s Theorem

**Green’s Theorem** relates a line integral around a simple closed curve $ C $ to a double integral over the region $ D $ it encloses:

$$
\oint_C P\,dx + Q\,dy = \iint_D \left( \frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y} \right)\,dx\,dy
$$

### Example: Use Green’s Theorem to Compute a Circulation

Let $ \vec{F}(x, y) = (-y, x) $, and $ C $ is the unit circle $ x^2 + y^2 = 1 $:

````python
x, y = sp.symbols('x y')
P = -y
Q = x

# Compute partial derivatives
dQ_dx = sp.diff(Q, x)
dP_dy = sp.diff(P, y)
curl = dQ_dx - dP_dy

# Double integral over the unit disk
r, theta = sp.symbols('r theta')
jacobian = r  # Polar coordinates
integrand = curl.subs({x: r*sp.cos(theta), y: r*sp.sin(theta)}) * jacobian
area = sp.integrate(
    sp.integrate(integrand, (r, 0, 1)),
    (theta, 0, 2 * sp.pi)
)
print(area)  # Output: 2*pi
````

## 5. More Advanced Topics

### 5.1. Divergence and Curl

- **Divergence:** Measures the "outflowing-ness" of a vector field.
    $$
    \nabla \cdot \vec{F} = \frac{\partial P}{\partial x} + \frac{\partial Q}{\partial y}
    $$
- **Curl:** Measures the "rotation" at a point (in 2D, scalar-valued).
    $$
    \nabla \times \vec{F} = \frac{\partial Q}{\partial x} - \frac{\partial P}{\partial y}
    $$

#### Example: Compute Divergence and Curl

````python
div_F = sp.diff(P, x) + sp.diff(Q, y)
curl_F = sp.diff(Q, x) - sp.diff(P, y)
print("Divergence:", div_F)  # Output: 0
print("Curl:", curl_F)      # Output: 2
````
