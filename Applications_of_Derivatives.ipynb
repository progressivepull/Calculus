## Optimization

**Optimization** uses derivatives to find maxima or minima—like largest area, shortest path, etc.

### Example: Maximize the Area of a Rectangle with Fixed Perimeter

Suppose a rectangle has a perimeter of 20 meters. What dimensions maximize the area?

- Let width = x, length = y
- $2x + 2y = 20 \implies y = 10 - x$
- Area $A = x \cdot y = x(10-x)$

````python
from sympy import symbols, diff, solve

x = symbols('x')
A = x * (10 - x)
A_prime = diff(A, x)
critical_points = solve(A_prime, x)
print("Critical points:", critical_points)

# Second derivative test
A_double_prime = diff(A_prime, x)
for pt in critical_points:
    print(f"A''({pt}) = {A_double_prime.subs(x, pt)}")
````

Result: The area is maximized when $x = 5$, so the rectangle is a square.

## Related Rates

**Related rates** problems deal with how different quantities change with respect to time.

### Example: Inflating a Balloon

A spherical balloon’s radius increases at 2 cm/s. How fast is its volume increasing when the radius is 5 cm?

- Volume: $V = \frac{4}{3}\pi r^3$
- $ \frac{dV}{dt} = 4\pi r^2 \cdot \frac{dr}{dt} $

````python
from sympy import symbols, pi

r = symbols('r')
dV_dt = 4 * pi * r**2 * 2  # dr/dt = 2
rate_at_5 = dV_dt.subs(r, 5).evalf()
print("dV/dt when r=5:", rate_at_5, "cm^3/s")
````

Result: $dV/dt = 200\pi \approx 628.32\, \mathrm{cm}^3/\mathrm{s}$

````python
## Curve Sketching

**Curve sketching** uses derivatives to find critical points (max/min), inflection points, and general function shape.

### Example: Sketch $f(x) = x^3 - 3x^2 + 2$
````

````python
from sympy import symbols, diff, solve, lambdify
import numpy as np
import matplotlib.pyplot as plt

x = symbols('x')
f = x**3 - 3*x**2 + 2
f_prime = diff(f, x)
f_double_prime = diff(f_prime, x)

crit_pts = solve(f_prime, x)
infl_pts = solve(f_double_prime, x)

f_np = lambdify(x, f, 'numpy')
x_vals = np.linspace(-1, 4, 400)
plt.plot(x_vals, f_np(x_vals), label='f(x)')
plt.scatter([float(c) for c in crit_pts], [float(f.subs(x, c)) for c in crit_pts], color='red', label='Critical Points')
plt.scatter([float(i) for i in infl_pts], [float(f.subs(x, i)) for i in infl_pts], color='green', label='Inflection Points')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Curve Sketching with Python')
plt.grid(True)
plt.show()
````

## Motion

In physics, derivatives connect position, velocity, and acceleration.

### Example: Vertical Motion Under Gravity

A ball’s height after t seconds: $s(t) = 20t - 5t^2$. Find velocity, acceleration, and plot all.

````python
t = symbols('t')
s = 20*t - 5*t**2
v = diff(s, t)
a = diff(v, t)

print("Velocity formula:", v)
print("Acceleration formula:", a)

s_np = lambdify(t, s, 'numpy')
v_np = lambdify(t, v, 'numpy')
a_np = lambdify(t, a, 'numpy')

t_vals = np.linspace(0, 5, 200)
plt.plot(t_vals, s_np(t_vals), label='Position')
plt.plot(t_vals, v_np(t_vals), label='Velocity')
# plt.plot(t_vals, a_np(t_vals), label='Acceleration')
plt.xlabel('Time (s)')
plt.legend()
plt.title('Motion: Position, Velocity') #, Acceleration')
plt.grid(True)
plt.show()
````
