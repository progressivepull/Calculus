## What is Differentiation?

**Differentiation** finds the derivative of a function, which represents its rate of change. For a function $ f(x) $:

$$
f'(x) = \lim_{h\to0} \frac{f(x+h)-f(x)}{h}
$$

This gives the slope of the tangent line to the curve at any point $ x $.

---

## Computing Derivatives Symbolically

Let’s use [SymPy](https://sympy.org/) for symbolic differentiation.

### Example: Find the derivative of $ f(x) = x^3 + 2x^2 - x + 5 $

````python
from sympy import symbols, diff

x = Symbol('x')
f = x**3 + 2*x**2 - x + 5
f_prime = diff(f, x)
print("f(x):", f)
print("f'(x):", f_prime)
````

## Numerical Differentiation

For functions without a simple formula, we can approximate the derivative numerically.

### Example: Approximate the derivative of $ f(x) = e^x $ at $ x = 0 $

````python
import numpy as np

def f(x):
    return np.exp(x)

h = 1e-5
x0 = 0
numerical_deriv = (f(x0 + h) - f(x0)) / h
print("Approximate derivative at x=0:", numerical_deriv)
````

Compare this with the exact value (which is also $ e^0 = 1 $).

## Tangent Lines: Application of Derivatives

The **tangent line** to a curve at point $ x=a $ has slope $ f'(a) $ and passes through $ (a, f(a)) $:

$$
y = f'(a) \cdot (x - a) + f(a)
$$

### Visualizing a Tangent Line

Let’s plot $ f(x) = \sin(x) $ and its tangent at $ x = \pi/4 $:

````python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, sin, lambdify
from math import pi

x_sym = symbols('x')
f_sym = sin(x_sym)
f_prime_sym = diff(f_sym, x_sym)

# Create numerical functions
f_num = lambdify(x_sym, f_sym, 'numpy')
f_prime_num = lambdify(x_sym, f_prime_sym, 'numpy')

a = pi / 4
x_vals = np.linspace(0, np.pi, 400)
y_vals = f_num(x_vals)
slope = f_prime_num(a)
tangent = slope * (x_vals - a) + f_num(a)

plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label="f(x) = sin(x)")
plt.plot(x_vals, tangent, '--', label="Tangent at x=π/4")
plt.scatter([a], [f_num(a)], color='red')
plt.title("Function and Tangent Line at x=π/4")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
````

## Summary

- **Numerical differentiation** is useful for data or non-symbolic functions.
- Tangent lines illustrate the geometric meaning of derivatives and can be visualized with Matplotlib.
